```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn
import warnings
warnings.filterwarnings('ignore')
```

```python
#Reading the data
df = pd.read_csv('C:/Users/Public/Documents/Iris.csv')
```

```python
#first 5 rows of the data
df.head()
```

```python
df.drop(['Id'], axis = 1, inplace = True)
```

```python
#dimension of data
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#initializing Label Encoder
le = LabelEncoder()
df['Species'] = le.fit_transform(df['Species'])
df.head()
```

##### 0 - Species Iris-Setosa

##### 1 - Species Iris-Versicolor

##### 2 - Species Iris-Virginica

```python
#Separating independent and dependent variables
y = df['Species']
X = df.drop(['Species'], axis = 1)
```

```python
print(y)
print(X)
```

```python
#importing decision tree classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
#Create Decision tree function
dt_iris = DecisionTreeClassifier(random_state = 10) 
```

```python
dt_iris.fit(X,y)
```

```python
from sklearn import tree
```

```python
dt = tree.export_graphviz(dt_iris, out_file = 'Irt.dot')
```

```python
dt2 = tree.export_graphviz(dt_iris, out_file = 'irt.dot', feature_names = X.columns, max_depth = 6, filled = True, rounded = True, special_characters = True)
```

```python
!dot -Tpng irt.dot -o irt.png
```
